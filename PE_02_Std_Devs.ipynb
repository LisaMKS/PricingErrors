```python
import pandas as pd
import numpy as np
from datetime import datetime as dt, timedelta
import winsound

curr_date = dt.today().strftime('%Y%m%d')
```

# Part 1 Read Pricing Data CSV from PE_01

```python
%%time
df = pd.read_csv("legacy_conlin_contract20230206.csv", header = 0)
df.B1_EXVAT= df.B1_EXVAT.round(2)
df.B1_National_BUY_PRICE= df.B1_National_BUY_PRICE.round(2)
print(np.shape(df))
df.head(3)
```

## Remove nan Resus Conlin buy price, zero buy, zero sell

```python
df=df.loc[(df['CL_CONTR_CODE'].notnull())]
print(np.shape(df))
df=df.loc[(df['B1_EXVAT']!=0)]
print(np.shape(df))
df=df.loc[(df['CL_ORD_PRC_BND_1']!=0)]
print(np.shape(df))
df=df.drop(['DIST_CODE_x' ], axis=1)
df
```

## Add margin

```python
df['%Margin']=round(((df['B1_EXVAT']-df['CL_ORD_PRC_BND_1'])/df['B1_EXVAT']),3)
df
```

# Part 2 - Analyse

## 1. Create std dev summary

```python
std_df=df.groupby(['CL_CONTR_CODE', 'PG_ID','CHANNEL'])['%Margin'].std().reset_index().sort_values(by=['%Margin'],ascending=False)
# add PGID Name and Category
pgid_name_class= pd.read_excel("PGID_name_classific.xlsx")
std_df= pd.merge(std_df, pgid_name_class,  how='left', on=['PG_ID'])
#drop columns
cols = ['CL_CONTR_CODE', 'PG_ID', 'PG_NAME','CHANNEL', 'Category', '%Margin']
std_df = std_df[cols]
std_df.to_excel('std_df_'+curr_date+'.xlsx', index = False)  
std_df
```
