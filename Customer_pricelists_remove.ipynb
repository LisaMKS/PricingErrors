## The Functions

```python
def input_customer(code):   
    try:
        name_path = html_root+'/form/div[1]/div/div/div[1]/div/div/div/div[3]/div/div[2]/div/div/div/div/div/div/div/div[2]/div/div[1]/div/div/div/div[1]/div[3]/table/tbody/tr/td[1]/div/div/div[1]/div/div[2]/div/table/tbody/tr/td/div/div[2]/div/div/div/div/div/div[1]/div/div/div/div/div/div/div/div/div[2]/div/div/div/div/div/div/div/div/div[1]/div/div[2]/div/div/table/tbody/tr/td[1]/table/tbody/tr[1]/td[2]/table/tbody/tr/td[2]/table/tbody/tr/td/span/input'
        driver.find_element(By.XPATH,name_path).clear()
        driver.find_element(By.XPATH,name_path).send_keys(code)
        sleep(1)
    except:
        search_dropdown = html_root+'/form/div[1]/div/div/div[1]/div/div/div/div[3]/div/div[2]/div/div/div/div/div/div/div/div[2]/div/div[1]/div/div/div/div[1]/div[3]/table/tbody/tr/td[1]/div/div/div[1]/div/div[2]/div/table/tbody/tr/td/div/div[2]/div/div/div/div/div/div[1]/div/div/div/div/div/div/div/div/div[2]/div/div/div/div/div/div/div/div/div[1]/div/div[1]/table/tbody/tr/td[1]'
        sleep(0.5)
        driver.find_element(By.XPATH, search_dropdown).click()
        sleep(1)
        name_path = html_root+'/form/div[1]/div/div/div[1]/div/div/div/div[3]/div/div[2]/div/div/div/div/div/div/div/div[2]/div/div[1]/div/div/div/div[1]/div[3]/table/tbody/tr/td[1]/div/div/div[1]/div/div[2]/div/table/tbody/tr/td/div/div[2]/div/div/div/div/div/div[1]/div/div/div/div/div/div/div/div/div[2]/div/div/div/div/div/div/div/div/div[1]/div/div[2]/div/div/table/tbody/tr/td[1]/table/tbody/tr[1]/td[2]/table/tbody/tr/td[2]/table/tbody/tr/td/span/input'
        driver.find_element(By.XPATH,name_path).clear()
        driver.find_element(By.XPATH,name_path).send_keys(code)
        sleep(1)
           
def search():
        search_customer_button=html_root+'/form/div[1]/div/div/div[1]/div/div/div/div[3]/div/div[2]/div/div/div/div/div/div/div/div[2]/div/div[1]/div/div/div/div[1]/div[3]/table/tbody/tr/td[1]/div/div/div[1]/div/div[2]/div/table/tbody/tr/td/div/div[2]/div/div/div/div/div/div[1]/div/div/div/div/div/div/div/div/div[2]/div/div/div/div/div/div/div/div/div[1]/div/div[2]/div/table/tbody/tr/td/table/tbody/tr/td[3]/button'
        driver.find_element(By.XPATH, search_customer_button).click()
        sleep(2)

#uses partial link    
def open_customer(code): 
    link_text = str(code + ' -')
    driver.find_element(By.PARTIAL_LINK_TEXT,link_text).click()
    sleep(3)

def read_table(table):
    rows = table.find_elements(By.TAG_NAME,'tr')
    read_table.row_elements = []
    read_table.row_element_names = []
    read_table.pricelist_found = False
    for row in rows:
        if row.text != '':
            names = row.find_elements(By.TAG_NAME,'td')
            read_table.row_element_names.append(names[2].text)
            read_table.row_elements.append(row)   
    
def search_for_pricelist(row_element_names, row_elements,pricelist):
    for i,name in enumerate(row_element_names):
        read_table.pricelist_found = False
        if pricelist == name:
            print('Delete')      
            row_elements[i].click()
            delete_pricelist()
            read_table.pricelist_found = True
            break
    

def scrolltop(number):
    scroll_path = html_root+'/form/div[1]/div/div/div[1]/div/div/div/div[3]/div/div[2]/div/div/div/div/div/div/div/div[2]/div/div[1]/div/div/div/div[1]/div[3]/table/tbody/tr/td[1]/div/div/div[1]/div/div[2]/div/table/tbody/tr/td/div/div[3]/div/div/div/div/div/div[1]/div/div/div/div/div/div/div/div/div[2]/div/div/div/div/div/div/div/span[2]/div/table/tbody/tr/td/div/div[1]/div/div/span/span[2]/div/div[2]/div/div[2]/div/span/div[1]/div[2]/div/div[6]'
    fBody = driver.find_element(By.XPATH,scroll_path)
    scroll = 0
    while scroll < number:  # this will scroll 3 times
        driver.execute_script('arguments[0].scrollTop = arguments[0].scrollTop + arguments[0].offsetHeight;',fBody)
        scroll += 1
        # add appropriate wait here, of course. 1-2 seconds each
        sleep(2.5)

def scrolldown():
    scroll_path = html_root+'/form/div[1]/div/div/div[1]/div/div/div/div[3]/div/div[2]/div/div/div/div/div/div/div/div[2]/div/div[1]/div/div/div/div[1]/div[3]/table/tbody/tr/td[1]/div/div/div[1]/div/div[2]/div/table/tbody/tr/td/div/div[3]/div/div/div/div/div/div[1]/div/div/div/div/div/div/div/div/div[2]/div/div/div/div/div/div/div/span[2]/div/table/tbody/tr/td/div/div[1]/div/div/span/span[2]/div/div[2]/div/div[2]/div/span/div[1]/div[2]/div/div[6]'
    fBody = driver.find_element(By.XPATH,scroll_path)   
    driver.execute_script('arguments[0].scrollTop = arguments[1]',fBody)
    sleep(2.5)

def move_national(pricelist):
    try:
        your_link = driver.find_element(By.LINK_TEXT,pricelist)
        actionChains.context_click(your_link).perform()
        sleep(1)                
    except:
        scrolltop(5)
        #ready = input('Scroll to the pricelist. Then click in the box below and hit enter')
        your_link = driver.find_element(By.LINK_TEXT,pricelist)
        actionChains.context_click(your_link).perform()
        sleep(1.2)
        
def delete_pricelist():
    delete_button=html_root+'/form/div[1]/div/div/div[1]/div/div/div/div[3]/div/div[2]/div/div/div/div/div/div/div/div[2]/div/div[1]/div/div/div/div[1]/div[3]/table/tbody/tr/td[1]/div/div/div[1]/div/div[2]/div/table/tbody/tr/td/div/div[3]/div/div/div/div/div/div[1]/div/div/div/div/div/div/div/div/div[2]/div/div/div/div/div/div/div/span[2]/div/table/tbody/tr/td/div/div[1]/div/div/span/span[2]/div/div[2]/div/div[2]/div/span/div[1]/div[1]/div/div/table/tbody/tr/td[2]/div/div[1]/div[1]/table/tbody/tr/td[1]'
    driver.find_element(By.XPATH,delete_button).click()
    sleep(4)
    yes_button=html_root+'/form/div[2]/div[2]/div[1]/div[1]/table/tbody/tr/td/div/div/table/tbody/tr[3]/td[2]/table/tbody/tr/td[1]/table/tbody/tr/td[3]/button'
    driver.find_element(By.XPATH,yes_button).click()
    sleep(1.5)
    
                       
def save_and_close():
    save_close_button=html_root+'/form/div[1]/div/div/div[1]/div/div/div/div[3]/div/div[2]/div/div/div/div/div/div/div/div[2]/div/div[1]/div/div/div/div[1]/div[3]/table/tbody/tr/td[1]/div/div/div[1]/div/div[2]/div/table/tbody/tr/td/div/div[3]/div/div/div/div/div/div[1]/div/div/div/div/div/div/div/div/div[1]/div/div[2]/div/div/div/div/div/table/tbody/tr/td[2]/div/div/div[1]/div/div[1]/div[3]/div/div[1]/div[1]/table/tbody/tr/td[5]'
    driver.find_element(By.XPATH,save_close_button).click()
    sleep(2.5)

def play_tune_tg():  
    length = 400
    length2 = 200
    winsound.Beep(250,length)
    winsound.Beep(370,length)
    winsound.Beep(370,length)
    
```
