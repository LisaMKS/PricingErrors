# Part 2 - Create Summary Files

## Group by PGID and NPC Sales

```python
libra2023_YTD=libra_df.groupby(['NPC','PG_ID']).agg({'SumOfSUPPLIED_INDIVIDUAL_UNITS':'sum','SumOfSALES':'sum'}).reset_index()
libra2023_YTD=libra2023_YTD.rename(columns={"SumOfSUPPLIED_INDIVIDUAL_UNITS":"YTD_SUPPLIED_UNITS","SumOfSALES":"YTD_SALES"})
libra2023_YTD
```

```python
libra2023_YTD_Sales=libra2023_YTD.loc[libra2023_YTD['YTD_SUPPLIED_UNITS']!=0]
libra2023_YTD_Sales.to_csv('libra2023_YTD_Sales.csv', index = False)
print(np.shape(libra2023_YTD_Sales))
```

## Four weeks Sales

```python
four_Weeks=libra_df.loc[libra_df['NHS_WEEK_NO']>=40]
```

```python
libra2023_four_Weeks=four_Weeks.groupby(['NPC','PG_ID']).agg({'SumOfSUPPLIED_INDIVIDUAL_UNITS':'sum','SumOfSALES':'sum'}).reset_index()
libra2023_four_Weeks=libra2023_four_Weeks.rename(columns={"SumOfSUPPLIED_INDIVIDUAL_UNITS":"4WKS_SUPPLIED_UNITS","SumOfSALES":"4WKS_SALES"})
libra2023_four_Weeks
```

```python
libra2023_four_Weeks_sales=libra2023_four_Weeks.loc[libra2023_four_Weeks['4WKS_SUPPLIED_UNITS']!=0]
libra2023_four_Weeks_sales.to_csv('libra2023_four_Weeks_Sales.csv', index = False)
print(np.shape(libra2023_four_Weeks_sales))
```
