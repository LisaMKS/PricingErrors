```python
%%time
df = pd.read_csv("price_data_20230210.csv", header = 0)
df.B1_EXVAT= df.B1_EXVAT.round(2)
df.B1_National_BUY_PRICE= df.B1_National_BUY_PRICE.round(2)
df['EFFECTIVE_DATE'] =  pd.to_datetime(df['EFFECTIVE_DATE'])
df['CP_EFFECTIVE_DATE'] =  pd.to_datetime(df['CP_EFFECTIVE_DATE'])
print(np.shape(df))
df.head(3)
```

## Exclude Zero buy, sell

```python
price_data=df
print(np.shape(price_data))
price_data=price_data.loc[price_data['B1_EXVAT']>.01]
print(np.shape(price_data))
price_data=price_data.loc[price_data['B1_National_BUY_PRICE']>.01]
print(np.shape(price_data))
price_data=price_data.loc[price_data['CL_ORD_PRC_BND_1']!=0]
print(np.shape(price_data))
```

```python
national = price_data.loc[price_data['PG_ID']==1]
national
```

```python
price_data['Nat_Margin_pc']=round(((price_data['B1_EXVAT']-price_data['B1_National_BUY_PRICE'])/price_data['B1_EXVAT']*100),1)
price_data['Reg_Margin_pc']=round(((price_data['B1_EXVAT']-price_data['CL_ORD_PRC_BND_1'])/price_data['B1_EXVAT']*100),1)
```

## Calc medians

```python
nat_medians=price_data.groupby(['CL_CONTR_CODE', 'PG_ID'])['Nat_Margin_pc'].median().reset_index(name='nat_median')
nat_medians.sort_values(by=['PG_ID'],ascending=True)
```

```python
national = nat_medians.loc[nat_medians['PG_ID']==1]
national
```

```python
reg_medians=price_data.groupby(['CL_CONTR_CODE', 'PG_ID'])['Reg_Margin_pc'].median().reset_index(name='reg_median')
reg_medians
```

## Merge Price Data with Nat and Reg Medians, Create Columns with Diff

```python
prices_nat_medians = pd.merge(price_data,nat_medians,on=['PG_ID','CL_CONTR_CODE'],how='left')
prices_medians = pd.merge(prices_nat_medians,reg_medians,on=['PG_ID','CL_CONTR_CODE'],how='left')
```

```python
prices_medians['Nat_Median_diff']=round((prices_medians['Nat_Margin_pc']-prices_medians['nat_median']),1)
prices_medians['Reg_Median_diff']=round((prices_medians['Reg_Margin_pc']-prices_medians['reg_median']),1)
```

```python
df=prices_medians
```
